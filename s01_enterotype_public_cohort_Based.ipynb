```r
###-----------------------------------------------------------------------------------------------------------------------
##### get PCoA and clustering
###-----------------------------------------------------------------------------------------------------------------------
bact= prop.table(feat.all_adjusted,margin = 2) %>% t %>% as.data.frame
bact[1:4,1:4]
md= meta.crc %>% column_to_rownames('motuID')
###
head(md)
table(md$Study)
###
input= cbind(md, bact[rownames(md),])
var_bact= colnames(bact)
###----------------------------------------------------------------
## filter
###----------------------------------------------------------------
head(tax)
taxtmp= tax
rownames(taxtmp)= paste0(taxtmp$Species, '.[',rownames(taxtmp),']')
head(taxtmp)
###
all(colnames(bact) %in% rownames(taxtmp))
###
taxtmp= taxtmp[colnames(bact),]

OTU = otu_table(bact %>% t, taxa_are_rows = TRUE)
TAX = tax_table(taxtmp %>% as.matrix)
samples = sample_data(md)
carbom <- phyloseq(OTU, TAX, samples)
#
carbom_genus= tax_glom(carbom,rank_names(carbom)[6])
#carbom= carbom_genus
############################filter
minTotRelAbun = 1e-5
x = taxa_sums(carbom)
keepTaxa = taxa_names(carbom)[which((x / sum(x)) > minTotRelAbun)]
prunedSet = prune_taxa(keepTaxa, carbom)
####
minTotRelAbun= 0
prunedSet = filter_taxa(carbom, function(x) sum(x) > minTotRelAbun, TRUE)
############################
prunedSet@tax_table %>% rownames() %>% length()
###
#var_bact= colnames(bact)
var_bact= prunedSet@tax_table %>% rownames()
length(var_bact)
###
bact= otu_table(prunedSet) %>% t %>% as.data.frame
dim(bact)
###
bact[1:4,1:4]
input= cbind(md, bact[rownames(md),])
###
## setting parameters
## Evaluate number of kmeans cluster
max_dim <- 2
set.seed(42)
k <- 3
dismethod= 'gower'
###
## ----------------------------------------------------------------
## PCoA analysis cluster
## ----------------------------------------------------------------
# Compute distance matrix with gower
##
distance <- vegdist(input %>% 
                      #column_to_rownames("Sample_ID") %>%
                      select(all_of(var_bact)),
                    #na.rm = T,
                    method = 'gower')
distance <- sqrt(distance) # square-root transform to get positive PCoA eigen values

#' Compute PCoA
pcoa <- cmdscale(distance, k = nrow(input)-1, eig = TRUE)
## ----------------------------------------------------------------
## Kmeans hierarchical clustering on PCoA1 and PCoA2
## ----------------------------------------------------------------
#' Extract sample coordinates
data <- pcoa$points %>%
  as.data.frame.matrix() %>%
  rename_with(~ gsub("V", "PCoA", .x, fixed = TRUE))
##
##
optimal_cluster <- fviz_nbclust(data[, 1:max_dim], kmeans)
clust <- hkmeans(data[ , 1:max_dim], k = k)
#' Merge clusters with clinical data
samples <- data %>%
  rownames_to_column("Sample_ID") %>%
  left_join(md, by = "Sample_ID") #%>% 
#left_join(alpha, by = "Sample_ID") 
samples$clust = as.factor(clust$cluster)
head(samples)
###
samples$response_bin= samples$response_bin %>% factor(levels = c('responder','non_responder'))
samples$age_bin= ifelse(samples$age >60, 'old','young')
```

```r
## --------------------------------------------------------------------------------------------------------------------------------
## DA analysis using wilcox + conhendis methods
## --------------------------------------------------------------------------------------------------------------------------------
## *****************   load function ******************************************
library(colorspace)
library(cols4all)
library(effectsize)
#########################
#########################
#' @description
#' Description.
#'
#' @param group_A_values A param.
#' @param group_B_values A param.
#'
#' @return An output.
#'
#' @noRd
cohens_d <- function(group_A_values, group_B_values) {
  n_A <- length(group_A_values)
  n_B <- length(group_B_values)
  mu_A <- mean(group_A_values)
  mu_B <- mean(group_B_values)
  var_A <- var(group_A_values)
  var_B <- var(group_B_values)
  pooled_sd <- sqrt(((n_A - 1) * var_A + (n_B - 1) * var_B) / (n_A + n_B - 2))
  cohen_d <- (mu_A - mu_B) / pooled_sd
  cohen_d
}
##########################
##########################
#' @description
#' Description.
#'
#' @param consistencies_matrix Either your CompassData instance's reaction_consistencies matrix, or its metareaction_consistencies matrix, depending on whether you want a table whose rows correspond to reactions or metareactions.
#' @param group_A_cell_ids A character vector containing the IDs of the cells that constitute group A.
#' @param group_B_cell_ids A character vector containing the IDs of the cells that constitute group B.
#' @param for_metareactions Whether the first argument is your reaction_consistencies matrix or your metareaction_consistencies matrix. This argument doesn't affect the contents of the returned tibble, but merely ensures that its column names are appropriate.
#' @param ... Unused.
#'
#' @return A tibble, where each row represents a Wilcoxon rank-sum test for whether a reaction or metareaction achieves a higher consistency among the group A cells than among the group B cells. It has the following columns: reaction_id or metareaction_id, wilcoxon_statistic, cohens_d, p_value, and adjusted_p_value.
#'
#' @importFrom magrittr %>% %<>%
conduct_wilcoxon_test = function(consistencies_matrix, group_A_cell_ids, group_B_cell_ids, ..., for_metareactions = TRUE) {
  if (0 < length(intersect(group_A_cell_ids, group_B_cell_ids))) {
    message("Groups A and B are not mutually exclusive. Continuing anyways ...")
  }
  group_A_values_per_metareaction <-
    consistencies_matrix[group_A_cell_ids, ] %>%
    #t() %>%
    #tibble::as_tibble(rownames = self$settings$cell_id_col_name) %>%
    #dplyr::right_join(
    #  tibble::tibble(!!self$settings$cell_id_col_name := group_A_cell_ids),
    #  by = self$settings$cell_id_col_name
    #) %>%
    as.data.frame()
  group_B_values_per_metareaction <-
    consistencies_matrix[group_B_cell_ids,] %>%
    #t() %>%
    #tibble::as_tibble(rownames = self$settings$cell_id_col_name) %>%
    #dplyr::right_join(
    #  tibble::tibble(!!self$settings$cell_id_col_name := group_B_cell_ids),
    #  by = self$settings$cell_id_col_name
    #) %>%
    as.data.frame()
  metareaction_ids <- colnames(consistencies_matrix)
  wilcoxon_results <-
    purrr::map_dfr(
      metareaction_ids,
      function(metareaction_id) {
        group_A_values <- group_A_values_per_metareaction[,metareaction_id] %>% as.numeric
        group_B_values <- group_B_values_per_metareaction[,metareaction_id] %>% as.numeric
        wilcoxon_result_obj <- wilcox.test(group_A_values, group_B_values)
        wilcoxon_result_tbl <- data.frame(
          metareaction_id = metareaction_id,
          wilcoxon_statistic = wilcoxon_result_obj$statistic,
          cohens_d = cohens_d(group_A_values, group_B_values),
          p_value = wilcoxon_result_obj$p.value,
          stringsAsFactors = FALSE
        )
        wilcoxon_result_tbl
      }
    ) %>%
    tibble::as_tibble() %>%
    dplyr::mutate(adjusted_p_value = p.adjust(dplyr::pull(., p_value), method = "BH"))
  if (!for_metareactions) {
    wilcoxon_results %<>% dplyr::rename(reaction_id = metareaction_id)
  }
  wilcoxon_results
}
```
